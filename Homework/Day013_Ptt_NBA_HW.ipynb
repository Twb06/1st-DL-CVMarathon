### ① 印出最新文章的「作者」「標題」「時間」

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.ptt.cc/bbs/NBA/index.html"
resp = requests.get(url).text
soup = BeautifulSoup(resp, "html5lib")

print(soup.prettify())
```

### ② 印出第一頁所有文章的「作者」「標題」「時間」

```python
posts_list = soup.find_all(class_ = "r-ent")

for post in posts_list:
    author = post.find(class_ = "author").text
    title = post.find(class_ = "title").find("a").text
    date= post.find(class_ = "date").text
    
    print(author, title, date)
```

### ③ 試著爬爬看其他版的文章

```python
# 八卦版
url = "https://www.ptt.cc/bbs/Gossiping/index.html"
cookies = {"over18" : "1"}
resp = requests.get(url, cookies = cookies).text
soup = BeautifulSoup(resp, "html5lib")

print(soup.prettify())
```

```python
posts_list = soup.find_all(class_ = "r-ent")

for post in posts_list:
    author = post.find(class_ = "author").text
    
    title_html = post.find(class_ = "title")
    if title_html.find("a") == None:
        continue
    else:
        title = title_html.find("a").text
        
    date= post.find(class_ = "date").text
    
    print(author, title, date)
```
