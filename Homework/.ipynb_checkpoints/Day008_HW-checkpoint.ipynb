### 1. Dcard 網址： https://www.dcard.tw/f

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.dcard.tw/f'

r = requests.get(url)
r.encoding = 'utf-8'
response = r.text
print(response[0:3000])
```

```python
print('Request 取回之後該怎麼取出資料，資料型態是什麼？ =>')

import json

#此例不是連接api，返回的是html格式，不是json可以解析的
#data = json.loads(response)
#直接讀取(r.text)的資料型態為str
print(type(response))
```

```python
print('為什麼要使用 BeatifulSoup 處理？處理後的型態是什麼？ => ')

#此例不是連接api，返回的是html格式，不是json可以解析的
#需要使用BeautifulSoup處理
soup = BeautifulSoup(response, "html5lib")

print(type(soup))
```

### 2. 知乎： https://www.zhihu.com/explore

```python
url = 'https://www.zhihu.com/explore'
r = requests.get(url)
r.encoding = 'utf-8'

print(r.text[0:600])
```

### 3. 觀察一下知乎回來的資料好像有點怪怪的，該怎麼解決？

```python
url = 'https://www.zhihu.com/explore'
#加入headers
headers = {"user-agent": "Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.130 Safari/537.36 OPR/66.0.3515.72"}

r = requests.get(url, headers = headers)
r.encoding = 'utf-8'
response = r.text
print(response[0:600])
```

```python
soup = BeautifulSoup(response, "html5lib")

print(soup.prettify())
```

```python
#抓取文章連結的標題文字
content_tag = soup.find_all('a', class_='ExploreSpecialCard-contentTag')
#content_tag返回的是list格式
for content in content_tag:
    print(content.text)
```
