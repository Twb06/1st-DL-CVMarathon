### 1. 根據範例使用 selenuim 取出 Hahow 網站中所有課程的資料，並用一個適合的資料結構儲存（Hint: 將所有資料存在一個變數當中取出）

```python
from selenium.webdriver.chrome import service
from selenium import webdriver
from bs4 import BeautifulSoup

# using opera driver
webDriverServer = service.Service('./Data/operadriver_win64/operadriver.exe')
webDriverServer.start()

capabilities = {
    'operaOptions': {
        'binary': 'C:/Program Files (x86)/Opera/launcher.exe'
    }
}

browser = webdriver.Remote(webDriverServer.service_url, webdriver.DesiredCapabilities.OPERA)

browser.get("https://hahow.in/courses")
```

```python
# 取得資料
html_source = browser.page_source
soup = BeautifulSoup(html_source, 'html5lib')
print(soup.prettify())
```

```python
courses_tags = soup.find_all("h4", class_ = "title marg-t-20 marg-b-10")
courses = [course.text for course in courses_tags]

print(courses)
```

```python
browser.close()
```

### 2. 剛剛的做法只能取出第一頁，試著取出所有頁面的資料。

```python
from selenium.webdriver.chrome import service
from selenium import webdriver
from bs4 import BeautifulSoup

import time

# using opera driver
webDriverServer = service.Service('./Data/operadriver_win64/operadriver.exe')
webDriverServer.start()

capabilities = {
    'operaOptions': {
        'binary': 'C:/Program Files (x86)/Opera/launcher.exe'
    }
}

browser = webdriver.Remote(webDriverServer.service_url, webdriver.DesiredCapabilities.OPERA)

browser.get("https://hahow.in/courses")
```

```python
# 取得資料
html_source = browser.page_source
soup = BeautifulSoup(html_source, 'html5lib')
print(soup.prettify())
```

```python
# 1. 先取出全部有幾頁
total_pages = soup.find("ul", class_ = "rc-pagination gbga9a-0 jYLVph").find_all("li")[-2].text
```

```python
# 2. 每頁都取出資料放入 courses
courses = []

for page in range(1, int(total_pages) + 1):
    url = "https://hahow.in/courses?page=" + str(page)
    browser.get(url)
    time.sleep(3)
    
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, 'html5lib')
    
    courses_tags = soup.find_all("h4", class_ = "title marg-t-20 marg-b-10")
    courses.extend([course.text for course in courses_tags])
```

```python
# 3. done!

print(courses)
```
